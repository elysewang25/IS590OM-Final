```python
def read_csv_input(input_file):
    res = []
    with codecs.open(os.path.join(input_file), 'r', 'ISO-8859-1') as f:
        #(x.replace('\0', '') for x in f)
        reader = csv.DictReader(f)

        for row in reader:
            if row['State'] != None and row['State'] != '' and row['Isglobal'] == 'us': #remove null value
                #res_dict = {}
                #dt = datetime.strptime(row['Time'], '%m/%d/%Y %H:%M')
                #res_dict['UFO Year'] = dt.year # reset format of UFO.year
                #res_dict['State'] = row['State']
                #res.append(res_dict)
                res.append(dict(row))

    return res
```

```python
def WriteDictToCSV(csv_file,csv_columns,dict_data):
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in dict_data:
            writer.writerow(data)
    return
```

## Combine data

Now we have several columns remain:  
UFO => {'State', 'Time', 'City', 'Form', 'Event Time(sec)', 'Description', 'X coordinate', 'Y coordinate'}
Population => {'State', 'Year from 1949 to 2013'}  
Climate => {'States','States Abbr.','Record Dates','Temperature Value', 'Temperature Rank', 'Temperature Departure from mean','Precipitation Value', 'Precipitation Rank', 'Precipitation Departure from mean'}  
  
  

However, there are still a few things that is needed to be done. For dataset Population, since all tuples' values in population are /k based, I will have to make all year.population values*1000.

```python
dictp = dfp.to_dict('records')
for data in dictres:
    for dp in dictp:
        if dp['States'].lower() == data['States'].lower():
            data['Population'] = int(dp[int(data['Record Dates'])]*1000) #all pp values*1000
            
dictres
```

In addition, for dataset UFO, I would like to delete a few columns include {'Isglobal', 'Event Time',  and 'Record Time'}.

```python
for du in dictu:
    dt = datetime.strptime(du['Time'], '%m/%d/%Y %H:%M')
    du_year = dt.year
    for data in dictres:
        if du['State'].lower() == data['States Abbr.'].lower() and du_year == int(data['Record Dates']):
            du['State Abbr.'] = data['States Abbr.']
            du['State'] = data['States']
            du['Temperature Value'] = data['Temperature Value']
            du['Temperature Rank'] = data['Temperature Rank']
            du['Temperature Departure from mean'] = data['Temperature Departure from mean']
            du['Precipitation Value'] = data['Precipitation Value']
            du['Precipitation Rank'] = data['Precipitation Rank']
            du['Precipitation Departure from mean'] = data['Precipitation Departure from mean']
            du['Population'] = data['Population']
            del du['Isglobal']
            del du['Event Time']
            del du['Record Time']
```
